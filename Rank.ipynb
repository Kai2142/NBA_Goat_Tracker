```python
import pandas as pd
```

```python
df = pd.read_csv("NBA_Goat_Tracker_Data.csv")
```

```python
df.info()
```

Rank PPG

```python
import pandas as pd

# Assuming df contains the DataFrame with 'Player Name' and 'PPG' columns
df_ppg_rank = df[['Player Name', 'PPG']].sort_values(by='PPG', ascending=False)

# Assign points based on ranking from 20 downwards, considering higher PPG values get more points
df_ppg_rank['Points'] = 21 - df_ppg_rank['PPG'].rank(method='min', ascending=False).astype(int)

# Add a column for rank based on PPG
df_ppg_rank['Rank'] = df_ppg_rank['PPG'].rank(method='min', ascending=False).astype(int)

# Display the Player Name, PPG, Rank, and Points for the top 20 players
df_ppg_rank.head(20)[['Rank','Player Name', 'PPG', 'Points']]
```
